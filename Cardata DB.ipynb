"4 . Dump all the data into car table \n",

```python
import csv
with open('car.data','r') as f:
    car_data=csv.reader(f,delimiter='\n')
    for i in car_data:
        print(i[0].split(','))
        lst=i[0].split(',')
        cursor.execute('insert into cardatasett.car values(%s,%s,%s,%s,%s,%s,%s)',lst)
mydb.commit()
```
